### Exercise 2: Bar Plot with Colors

```python

import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create bar plot
plt.bar(categories, values, color=colors)

# Customize plot
plt.title('Bar Plot of Categories')
plt.xlabel('Categories')
plt.ylabel('Values')

# Display the plot
plt.show()
```
