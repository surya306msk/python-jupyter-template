### Exercise 5: Subplot Layout

```python

import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 10)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
scatter_x = np.random.rand(10) * 10
scatter_y = np.random.rand(10) * 100
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart
axs[0, 0].plot(x, y, marker='o', color='b', linestyle='-')
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(categories, values, color=['blue', 'red', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple', marker='x')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140, 
              colors=['blue', 'red', 'green', 'orange'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
